```python
dataDat = pd.read_csv('galaxia_starbust_SED.txt', sep='\s+',
                    index_col=False)
print(dataDat)
```

```python
plt.plot(dataDat, color='orangered', alpha=0.65)
# Add this line to change the x-axis scale to logarithmic
plt.xscale('log')
plt.yscale('log')
# Graphic settings
plt.title(r'SED de la Galaxia')
plt.xlabel('$\lambda$ (Micras)')
plt.ylabel('$Enega$')
# plt.legend()
plt.grid(color='g', linestyle='dotted', linewidth=1)
plt.legend()
```

```python
dataDat.plot()
plt.show()
```

```python
x = dataDat['lam(micras)']

y = dataDat['energia(erg/sec/micras)']

plt.plot(x, y, label='Datos')
# Datos de la escala. Sacar el logaritmo a los datos
# Logaritmo de los datos
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
# Maximum value
xVal = x.max()
yVal = y.max()
print(xVal, yVal)
xPosit = x.idxmax()
yPosit = y.idxmax()
print(xPosit, yPosit)
```
